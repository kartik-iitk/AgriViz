```python
import pandas as pd
import difflib
```

```python
df = pd.read_parquet('../src/gradio_try/data/data.parquet')
target = pd.read_json('../src/gradio_try/data/states.json')
df.head()
```

```python
target = target['features']
target_states = list({d['id'] for d in target})
print(target_states)
```

```python
states = df['StateName'].unique()
print(states.to_list())
```

```python
def get_best_match(wrong_name):
  match = difflib.get_close_matches(wrong_name, target_states, n=1, cutoff=0.6)
  return match[0] if match else None

state_map = {state: get_best_match(state.lower()) for state in states}
print(state_map)
```

```python
df['StateName'].replace(state_map, inplace=True)
```

```python
df_clean = df.dropna(subset=['StateName'])
```

```python
STATES = df_clean['StateName'].unique().to_list()
```

```python
for s in STATES:
  target = pd.read_json(f"../src/gradio_try/data/statewise/{s}.json")
  target = target['features']
  s_dist = df_clean[df_clean['StateName'] == s]['DistrictName'].unique()
  target_s_dist = list({d['id'] for d in target})
  
  def get_best_match(wrong_name):
    match = difflib.get_close_matches(wrong_name, target_s_dist, n=1, cutoff=0.6)
    return match[0] if match else None
  
  dist_map = {
    dist: get_best_match(dist.lower()) if isinstance(dist, str) else None
    for dist in s_dist
  }
  df_clean['DistrictName'].replace(dist_map, inplace=True)
  df_clean.dropna(subset=['DistrictName'], inplace=True)
```
